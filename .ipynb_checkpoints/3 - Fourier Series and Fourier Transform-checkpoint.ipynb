```python
import numpy as np
global np
import scipy as sp
import scipy.signal as signal
import matplotlib.pyplot as plt
import IPython.display as ipd
from ipywidgets import interact
import sys
import wave
import warnings
warnings.filterwarnings('ignore')
sys.path.append("../backend/")
%matplotlib inline

def load_wav(filepath, t_start = 0, t_end = 2**32) :
    """Load a wave file, which must be 22050Hz and 16bit and must be either
    mono or stereo. 
    Inputs:
        filepath: audio file
        t_start, t_end:  (optional) subrange of file to load (in seconds)
    Returns:
        a numpy floating-point array with a range of [-1, 1]
    """
    wf = wave.open(filepath)
    num_channels, sampwidth, fs, end, comptype, compname = wf.getparams()
    
    # for now, we will only accept 16 bit files at 22k
    assert(sampwidth == 2)
    assert(fs == 22050)

    # start frame, end frame, and duration in frames
    f_start = int(t_start * fs)
    f_end = min(int(t_end * fs), end)
    frames = f_end - f_start

    wf.setpos(f_start)
    raw_bytes = wf.readframes(frames)
    # convert raw data to numpy array, assuming int16 arrangement
    samples = np.fromstring(raw_bytes, dtype = np.int16)

    # convert from integer type to floating point, and scale to [-1, 1]
    samples = samples.astype(np.float)
    samples *= (1 / 32768.0)

    if num_channels == 1:
        return samples

    elif num_channels == 2:
        return 0.5 * (samples[0::2] + samples[1::2])

    else:
        raise('Can only handle mono or stereo wave files') 

def make_cosine(a, k, ph, N) :
    n = np.arange(N)
    return a * np.cos( 2 * np.pi * k / N * n + ph)

def find_peaks(x, thresh = 0.2) :
    ''' finds peaks in 1D vector.
    Inputs:
        x: input vector
        thresh: relative threshold value. Discard peak whose value is lower than (thresh * max_peak_value).
    Output:
        np.array of index values of peaks
    '''

    x0 = x[:-2]   # x
    x1 = x[1:-1]  # x shifted by 1
    x2 = x[2:]    # x shifted by 2

    peak_bools = np.logical_and(x0 < x1, x1 > x2) # where x1 is higher than surroundings
    values = x1[peak_bools]                       # list of all peak values

    # find a threshold relative to the highest peak
    th = np.max(values) * thresh
    
    # filter out values that are below th
    peak_bools = np.logical_and(peak_bools, x1 > th)

    peaks = np.nonzero( peak_bools )[0] + 1       # get indexes of peaks, shift by 1
    return peaks
```

## Complex Exponentials as Frequencies

Centuries ago a mathematician named Euler created what is today known as the Euler identity (as well as creating half of mathematics). Euler's identity says the following:
$$e^{j \theta} = \cos(\theta) + j \sin(\theta)$$
Though some of you will have seen this before, it is important to stop for a second and note mentally that complex exponential functions are just sines and cosines.

## Fourier Series

We showed in pervious lectures that we can decompose all signals into even and odd components. We also showed from last lab that if we have two sine waves of different frequencies then they are uncorrelated. 
So, if we want to see how much 80hz content there is inside of a signal we could simply calculate 
$$\frac{1}{\pi}\int_{-\pi}^{\pi}f(t)\cos(2\pi \times 80 \times t) + \frac{1}{\pi}\int_{-\pi}^{\pi}f(t)\sin(2\pi \times 80 \times t)$$ where the left side is how much 80hz content is in the even part of the signal and the right side is how much is in the odd part of the signal. 
We can do a check for both of these at the same time by using $$e^{j \omega t} = \cos(\omega t) + j \sin(\omega t)$$ (remember the frequency detector from last time?)

So, let's start crafting a tool to generate a plot of frequency vs magnitude.
The first step is going to be to start figuring out the frequency content in the signal. 
So we can write:
$$f(t) = even + odd$$
$$f(t) = \frac{1}{\pi} \sum_{m=0}^{\infty}Even_{m} \cos(mt) + \frac{1}{\pi} \sum_{m=0}^{\infty} Odd_{m}\sin(mt)$$
where $Even_{m} = \int_{-\pi}^{\pi} f(t) \cos(mt)dt$ and $Odd_{m} = \int_{-\pi}^{\pi} f(t) \sin(mt)dt$.

```python
def fourierseries(an,bn,sr,a0=0):
    linspace = np.linspace(0.0,2*np.pi,10*sr)
    returnable = np.zeros(10*sr)
    returnable[0:] = a0/2
    if(len(an)!=len(bn)):
        print("an and bn have different lengths")
    for i in range(len(an)):
        returnable += an[i]*np.cos((i+1)*linspace) + bn[i] * np.sin((i+1)*linspace)
    return returnable
    
def bnsquarewave(n):
    if (n%2 != 0):
        return 1.0/(n)
    else:
        return 0
    
bned = [bnsquarewave(x) for x in range(1,500)]

fourier = fourierseries(np.zeros(len(bned)),bned,1000,a0=0)
plt.plot(fourier);
```

## Fourier Transform
Where as the fourier series allows us to generate approximations of a function using frequencies, what about generating frequencies using a function? We talked before about the frequency detector, but this time we are going to generalize it. Before we were able to detect individual frequencies by seeing how correlated $cos(\omega t)$ was with our signal $f(t)$. However, this can be generalized to take care of complex signals. The way we generalize to complex signals is that changing our detection method to be with $e^{ -j \omega t}$
Suddenly our detection method becomes 
$$ F(j\omega) = \int_{-\infty}^{\infty} f(t) e^{-j\omega t} dt$$
And this is the Fourier Transform. Seriously, that's it. Not as bad as you thought :D

## Eigenfunctions of LTI Systems

When we talk about vector spaces we consider vectors in those vector spaces. With an operator, $T$ on that vector space we say that given a vector $v$ with the property that $Tv = \lambda v$ where $v$ is called the eigenvector and $\lambda$ is the eigenvalue of that eigenvector.

We are going to apply the same idea to the space of inputs to LTI systems. Let's consider what happens to an LTI system when we input $x(t) = e^{j \omega t}$ into an LTI system $h(t)$.
$$y(t) = \int_{\tau} h(\tau)x(t-\tau) d\tau$$
$$y(t) = \int_{\tau} h(\tau)e^{j\omega (t-\tau)} d\tau$$
$$y(t) = e^{j\omega t}\int_{\tau} h(\tau) e^{-j \omega \tau} d\tau$$
$$y(t) = e^{j \omega t} H(j \omega)$$
since 
$$H(j \omega) = \int_{\tau} e^{- j \omega \tau}$$

So, complex exponentials are eigenfunctions of LTI systems.

The great utility of knowing that the eigenfunctions of LTI systems are complex exponentials is the following:
if the input into our system is $e^{j\omega t}$ then the output of the system is $$y(t) = e^{j\omega t} H(j\omega)$$. Since our input can be decomposed into the sum of weighted exponentials these can be put into the system and we can return the output easily.
For be more exact, for an input of $a_{1}e^{j\omega_{1}t} + a_{2}e^{j \omega_{2}t}$ we return an output of $$y(t) = a_{1}H(j\omega_{1})e^{j\omega_{1}t} + a_{2}H(j\omega_{2})e^{j\omega_{2}t}$$

# Discrete Fourier Transform

The continuous time Fourier transform is very useful as a teaching tool, but it does not allow us to understand but is slightly impractical. Note that we will be working with a lot of digital signals this IAP. The standard fourier transform makes a lot of sense for continuous time signals, but in discrete time we have a slightly different transform to deal with the discrete nature of the points.
The discrete fourier transform is defined as 
$$X(k) = \sum_{n=0}^{N-1} x(n)e^{-\frac{2 \pi j k n}{N}}$$
where $N$ is the length of $x$ and $k$ is the freqency (each index into the DFT is $\frac{1}{T}$ change in frequency from the previous index where $T$ is the time domain difference in samples of $x(n)$).
We shall look at how this graphically relates to the standard Fourier Transform with the following image:



First, note that the discrete fourier transform generated a periodic signal in the frequency domain. This is because we are only considering finitely make frequencies, instead of all possible frequencies. 

This is a demo that should cement together your understanding of how this DFT works. We are going to sum over all possible frequencies by iterating over the values of N.

```python
# periodically extend x:
N = 128
x = make_cosine(1.0, 7.5, 0, N)
y = np.tile(x, 2)

plt.figure()
plt.plot(y);
```

```python
# pull out Amp, K, and phase from FFT:
x_ft = np.fft.rfft(x)
x_amp = np.abs(x_ft) * 2 / N
x_phase = np.angle(x_ft)
cos_params = zip(x_amp, np.arange(len(x_amp)), x_phase)

# but now, sort them in order of amplitude (the "most important" ones first)

cos_params = sorted(cos_params, key = lambda x : x[0], reverse=True)

@interact(num_sines=(0, len(cos_params) - 1))
def reconstruct(num_sines=0):
    x = np.zeros(N)
    y = np.zeros(N)
    params_txt = ''
    for p in cos_params[:num_sines]:
        y = make_cosine(p[0], p[1], p[2], N)
        params_txt = 'A=%.2f k=%d, phi=%.2f' % p
        x += y
    plt.figure(figsize=(12,8))
    plt.subplot(2,1,1)
    plt.plot(y, 'r')
    plt.ylim(-1.5, 1.5);
    plt.text(10, 1, params_txt, fontsize=20)
    plt.subplot(2,1,2)
    plt.plot(x)
    plt.ylim(-1.5, 1.5)
    plt.title("sum of top %d sinusoids" % num_sines)
    plt.show();
```

There are several interesting things to note. First, the highest frequency that can be shown here is $\frac{N}{2}$.
Secondly, the y-axis grows in size with N.
Thirdly, for a fixed N when we change the number of samples in our cosine we don't obtain any more resolution on the DFT (the spike at 16hz does not change in width).
You may ask then, how do we obtain more resolution of our DFT? 
Let's derive how the frequency resolution of a DFT. 
We have $\frac{f_{s}}{2}$ as the frequency range that we can display, and $\frac{N}{2}$ possible bins which means that we have $\frac{f_{s}}{N}$ Hz per bin. 
So, based on this, if we are given a bin what frequency does this represent?

## Zero Padding
We are going to look at an interesting phenomena of the DFT, the fact that adding zeros to the beginning or end of our signal means that we can obtain finer grained frequency info. Let's look at a demo!

```python
zp_len = 512 - len(x)
y = np.concatenate((x, np.zeros(zp_len)))

plt.figure(figsize=(10,7))
plt.subplot(2,2,1)
plt.plot(x)
plt.title('$x(n), k=7.5, N=128$')
plt.subplot(2,2,2)
plt.stem(np.abs(np.fft.rfft(x)))
plt.xlim(0, len(x)/6.)
plt.title('$|X(k)|$')
plt.subplot(2,2,3)
plt.plot(y)
plt.title('$x(n), k=7.5$, zero padded to $N=512$')
plt.subplot(2,2,4)
plt.stem(np.abs(np.fft.rfft(y)))
plt.xlim(0, len(y)/6.)
plt.title('$|X(k)|$')
plt.show()
```

The reason why zero padding has this effect is because it creates a longer signal which means that you can use a longer DFT, resulting in more bins packed closer together. 

Here is a demo that you can use to play around with. zpf is a zero padding factor, how many zeros do we add relative to the size of our signal.
ni is the number of periods in the larger sinusoid

```python
c1 = make_cosine(1, 49.1, 0, 1024)
c2 = make_cosine(1, 53.2, 0, 1024)
x = c1 + c2
plt.plot(x)
plt.show()
```

```python
@interact(zpf=(1,8), ni=(0,3))
def zp_test(zpf=1, ni=0):
    sig_len = (128,256,512,1024)[ni]
    N = sig_len * zpf
    xw = np.concatenate((x[:sig_len], np.zeros(N-sig_len)))
    mft = np.abs(np.fft.rfft(xw))
    peaks = find_peaks(mft, .5)
    
    plt.figure(figsize=(10, 7))
    plt.subplot(2,1,1)
    plt.plot(xw)
    txt = '$N=%d$    $L_{sig} = %d$    $L_{zp} = %d$' % (N, sig_len, N - sig_len)
    plt.title(txt)
    plt.subplot(2,1,2)
    plt.title('$|X(k)|$')
    plt.plot(mft[0:int(N/8)])
    plt.show()
```

You may have noted from above that the sinewave that we sent through didn't result in a perfect spike. Why is this?
This is a demo to help explain why:

```python
@interact(k=(6.0,8.0))
def cos_and_dft(k = 7.5):
    N = 128
    x = make_cosine(1.0, k, 0, N)
    plt.figure()
    plt.subplot(1,2,1)
    plt.plot(x)
    plt.ylim(-1.5, 1.5)
    plt.title('$x(n)$')
    plt.subplot(1,2,2)
    plt.stem(np.abs(np.fft.rfft(x)))
    plt.ylim(0,65)
    plt.title('$|X(k)|$')
    plt.show()
```

We refer to this problem as spectral leakage.

The "problem" is that the frequency k=7.5 is not an integer and is thus not "perfectly capturable" by a DFT, since all values of k are integers.
Frequency information that would fall perfectly into a bin have suddenly been "smeared" across multiply bins. In a lot of instances this won't happen because we will restrict $k$ to being an integer, but in other instances this isn't a problem.

## Short Time Fourier Transform

Sometimes we care about how frequency changes over time within our signal. We currently don't posses the machinery to be able to display the changing frequency. This problem is best illustrated with this image where the left image is the signal, middle is the DFT, and right is what we want.



The best way to get around this constraint is by breaking the signal into chunks and then preforming the DFT on those chunks. Visually that looks like this:



We note that the procedure has windows applied to each bin. These windows help to prevent the spectral leakage problem that we noted above.

Another thing to note is that we overlap our windows so that we can make sure that all information in the time domain is captured.

```python
def make_chirp(k_start, k_end, N) :
    n = np.arange(N)
    dk = (k_end - k_start) / float(N)
    return np.sin(2*np.pi * (k_start + dk/2 * n) * n / N)

x = make_chirp(10, 400, 8192)
def window(x, w_start, win) :
    x_win = win * x[w_start:w_start+len(win)]
    win_zp = np.zeros(len(x))
    win_zp[w_start:w_start+len(win)] = win
    x_win_zp = np.zeros(len(x))
    x_win_zp[w_start:w_start+len(win)] = x_win

    return x_win, win_zp, x_win_zp

win_len = 1024
win_pos = np.array((0,1,2,3)) * win_len

@interact(p=(0,3), w=(0,2))
def sliding_window(p=0, w=0):
    # choose window:
    win = (np.ones(win_len), sp.signal.triang(win_len), np.hanning(win_len))[w]
    
    x_win, win_zp, x_win_zp = window(x, win_pos[p], win)
    
    plt.figure(figsize=(10,8))
    plt.subplot(3,1,1)
    plt.plot(x)
    plt.plot(win_zp, 'r', linewidth=2)
    plt.ylim(-1.5, 1.5)
    plt.xlim(0, 5000)

    plt.subplot(3,1,2)
    plt.plot(x_win_zp)
    plt.ylim(-1.5, 1.5)
    plt.xlim(0, 5000)

    plt.subplot(3,2,5)
    plt.plot(x_win)

    plt.subplot(3,2,6)
    plt.plot(np.abs(np.fft.fft((x_win))))
    plt.xlim(0, 60)
    plt.show()
```

## STFT Tradeoff Between Time Domain Precision and Frequency Domain Precision

The STFT has an inherent tradeoff between time domain and frequency domain that cannot be easily avoided. We will demonstrate this with a demo:

```python
def plotspectrogram(dpi=100,windowlength=256):
    cos10 = make_cosine(a=1.0,k=10,ph=0,N=1000)
    cos25 = make_cosine(a=1.0,k=25,ph=0,N=1000)
    cos50 = make_cosine(a=1.0,k=50,ph=0,N=1000)
    cos100 = make_cosine(a=1.0,k=100,ph=0,N=1000)
    totalcos = np.concatenate((cos10,cos25,cos50,cos100))
    window = np.hanning(windowlength)
    f,t,Sxx = sp.signal.spectrogram(x=totalcos,fs=22050,window=window)
    plt.figure(dpi=dpi)
    plt.pcolormesh(t,f,Sxx)
    plt.show()
interact(plotspectrogram,dpi=(70,300),windowlength=[128,256,512,1024]);
```

So note that with the increased windowlength we can increase our precision in the frequency domain, but we really lose out on how precise we can be in the time domain.

## Sidenote: Relationship with Uncertainty Principle
The heisenberg uncertainty principle says that $$ \sigma_{x} \sigma_{p} \geq \frac{h}{4 \pi}$$ where $h$ is Planck's constant and $\sigma_{x}$ and $\sigma_{p}$ are the standard deviations of the position and momentum respectively.

The reason why this relates to this course is that one of the principles that the uncertainty principle is attempting to convey is that of finite information and the tradeoff between two measurable states. A version we could use this this course is $$\sigma_{t} \sigma{f} \geq \text{information of signal}$$
where $\sigma_{t}$ and $\sigma_{f}$ are the standard deviation of the measurements in time and frequency respectively.
